```python
import numpy as np
import matplotlib.pyplot as plt

from doggofier.dataset import DogsDataset
```

The Tsinghua Dogs is an image dataset for fine-grained classification of dog breeds. It includes 130 categories and over 70,000 images, 65% of them are collected from people's real life. Each category contains at least 200 images and a maximum of 7,449 images. In comparison to similar datasets, it has more categories and more carefully chosen images for each category. Additionally, the dataset is annotated with bounding boxes of the dog's body and head in each image, which can be used for object detection and localization problems. Here we present an exploaration of training part of the dataset.

```python
dataset = DogsDataset('data')

ids = len(dataset)
sample = dataset[np.random.choice(ids)]

sample
```

```python
categories = dataset.get_categories()
```

## Plot a sample image

```python
plt.imshow(sample[0])
plt.title(categories[sample[1]])
plt.show()
```

## Distribution of categories

```python
from collections import Counter, OrderedDict

cnt = dict(Counter(categories[sample[1]] for sample in dataset))
sorted_tuples = sorted(cnt.items(), key=lambda item: item[1])
sorted_cnt = OrderedDict({breed: count for breed, count in sorted_tuples})

sorted_cnt
```

```python
plt.figure(figsize=(12, 6))
plt.bar(list(sorted_cnt.keys())[:-21:-1],
        list(sorted_cnt.values())[:-21:-1],
        edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```
